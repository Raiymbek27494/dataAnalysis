```python
bankCust = pd.read_csv("Bank Customer Churn Prediction.csv")
df1 = pd.DataFrame.copy(bankCust[bankCust['active_member'] == True])
df1 = df1[(df1['balance']>0) & (df1['credit_card'] == True) & (df1['country'] == 'Germany') & (df1['estimated_salary'] > (1584 * 12))]
```

```python
#1-2
fig, ax = plt.subplots(figsize=(30, 12))
ax = sns.boxplot(y = "gender", x = "estimated_salary" , data=df1, color='tab:blue')
ax.tick_params(rotation=90, labelsize=18)
ax = sns.stripplot(y = "gender", x = "estimated_salary" ,data=df1, color="tab:red") 
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
```

```python
mask = np.array(Image.open('usa.jpeg'))
wc = WordCloud(stopwords=STOPWORDS,
               mask=mask, background_color="white",
               max_words=2000, max_font_size=256,
               random_state=42, width=mask.shape[1],
               height=mask.shape[0])
wc.generate(text)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
