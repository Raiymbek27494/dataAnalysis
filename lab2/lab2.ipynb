```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
rc = {'figure.figsize':(10,5),
      'axes.facecolor':'white',
      'axes.grid' : True,
      'grid.color': '.8',
      'font.family':'Times New Roman',
      'font.size' : 15}
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.rcParams.update(rc)
```

```python
goog = pd.read_csv('GOOG.csv', low_memory=False, parse_dates=[0])
goog.info()
```

```python
#1
goog['Date'], goog['High'], goog['Low']
```

```python
#2
nparray = np.array([goog.Date, goog.High, goog.Low])
nparray.shape, type(nparray)
```

```python
df1 = pd.DataFrame.copy(goog[['Date', 'High', 'Low', 'Open']])
df1.info()
```

```python
#3-5
fig1, ax1 = plt.subplots()
ax1.plot(df1.Date, df1.High)
ax1.plot(df1.Date, df1.Low, color='red')
ax1.set_title('Some simple plot')
ax1.set_xlabel('Date')
ax1.set_ylabel('Price')
ax1.fill_between(df1.Date, df1.High, df1.Low, color = 'lightgreen')
ax1.legend(['High','Low'])
plt.tight_layout()
plt.show()
```

```python
#6
fig2, ax2 = plt.subplots()
print(ax2)

ax2ins1 = ax2.inset_axes([.08, .75, .4, .2])
ax2ins1.plot(df1.Date, df1.High)
ax2ins1.set_ylabel('High_Price')
ax2ins1.legend('High')

ax2ins2 = ax2.inset_axes([.6, .11, .4, .2])
ax2ins2.plot(df1.Date, df1.Low, color='red')
ax2ins2.set_ylabel('Low_Price')
ax2ins2.legend('Low')

ax2.plot(df1.Date, df1.Open, color='lightgreen')
ax2.set_xlabel('Date')
ax2.set_ylabel('Open_Price')
ax2.legend('Open', loc='upper right')
plt.show()
```
